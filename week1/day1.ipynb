```python
system_prompt = """
You are a helpful assistant that analyzes the content of a website, and provides a short snarky, humorurs summary, 
ignoring text that might be naviagation related.
Respond in Markdown. Do not wrap markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt = """
Here is a contents of a website.
Provide a short summary of this website in a markdown.
If it includes news and announcements, then summarize these too.
"""
```

```python
messages = [
    {"role": "system" ,"content":"You are a snarky assistant."},
    {"role": "user" ,"content":"What is 2+2?"}
]
```

```python
response = openai.chat.completions.create(model="gpt-4.1-nano",messages=messages)
print(response.choices[0].message.content)
```

```python
def messages_for(website): 
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website},
    ]
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
# SUMMARIZER USING OLLAMA MODEL

def summarizer_Ollama(url):
    website = fetch_website_contents_farhan(url)
    response = ollama.chat.completions.create(
        model="deepseek-r1:1.5b", 
        messages=messages_for(website=website)
        )
    return display(Markdown(response.choices[0].message.content))
```

```python
# SUMMARIZER USING FRONTIER MODEL


def summarizer_Frontier(url):
    website = fetch_website_contents_farhan(url)
    response = openai.chat.completions.create(
        model="gpt-5.5", messages=messages_for(website=website)
    )
    return display(Markdown(response.choices[0].message.content))
```

```python
summarizer_Ollama("https://cnn.com/")
```

```python
summarizer_Frontier("https://cnn.com/")
```
